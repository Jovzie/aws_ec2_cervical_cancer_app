**Cervical Cancer: HPV, Other Risk Factors & Prevention

According to the Center for Disease Control (CDC), each year, about 4000 women die from cervical cancer, while 16,000 new diagnoses are made annually in America alone (Cervical Cancer Statistics | CDC). Globally, cervical cancer killed about 342,000 women in 2020 (Cervical cancer (who.int)). Almost all of the cervical cancer is caused by very specific “high-risk” strains (type 16, 18) of human papillomavirus (HPV) (Cervical cancer (who.int)). HPV is asymptomatic and it may be transmitted sexually or sometimes, via skin-to-skin contact of some parts of the body. In fact, HPV is the most commonly transmitted STI (i.e., sexually transmitted infection) (Cervical Cancer Statistics | CDC). Some estimates suggest that most people will have been exposed to some strain of HPV during their lifetime, while about half of the people in the population are currently infected (i.e., they are likely to pass them on to current sexual partners) (Cervical cancer (who.int)). Most people will not have been aware of having contracted the virus.  Interestingly, of those ever infected, most will have “cleared” the virus via the functions of the immune system over the span of two years. However, sometimes, the virus lingers and if it does, it is possible for it to cause cervical cancer. Several other risk factors to cervical cancer have been identified: smoking, having HIV, having given birth to more than 3 children, a higher number of sexual partners, and the use of some birth control methods.

**Preventative Measures

The best preventative measure for cervical cancers caused by HPV is vaccination, which is very effective (Cervical Cancer Statistics | CDC). Testing for HPV on STI panels is important in knowing one’s HPV status. Another extremely important preventative measure is regular screenings, including biopsies of cervical cells taken during pap smears which may show irregular cell growth. Early detection is important in treating cancer. Since many people, especially men who may infect women, will not get tested for HPV on STI panels, it is important to bring awareness to the risk factors and preventative measures associated with cervical cancer. It would be useful to come up with a machine learning model to predict cancer/not cancer based on multiple risk factors.

```python
#identifying numeric columns
numeric_columns = [
    "Age", "STDs: Number of diagnosis", "Dx:Cancer", "Dx:CIN", "Dx:HPV", "Dx", "Hinselmann", "Schiller", "Citology", "Biopsy"
]
print(numeric_columns)
```

```python
#dropping numeric columns from above - unary operation (opposite of all numeric is the object columns)
bool_columns = data.drop(columns=numeric_columns).columns
bool_columns
```

```python
#making sure all data are numeric in order to be able to impute (first nan)
data[bool_columns] = data[bool_columns].apply(pd.to_numeric, errors="coerce")

#filling in missing values (then filled in)
data[bool_columns].fillna(np.nan, inplace=True)
```

```python
#imputing missing values with SimpleImputer
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median") #chose median as data are skewed and mean doesn't work for categorical data
imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=list(data.columns))
imputed_data
```

```python
#checking out range of age variable
print(imputed_data["Age"].max())
print(imputed_data["Age"].min())
```

```python
#splitting groups according to age
def age_category(age):
    if age < 12:
        return "Child"
    elif age < 20:
        return "Teen"
    elif age < 30:
        return "20's"
    elif age < 40:
        return "30's"
    elif age < 50:
        return "40's"
    elif age < 60:
        return "50's"
    elif age < 70:
        return "70's"
    else:
        return "70+"

imputed_data["Age"] = imputed_data["Age"].astype(np.uint8) # coerce values to low precision integer for categorical data

#creating new variable Age_Category
imputed_data["Age_Category"] = imputed_data["Age"].apply(age_category)
imputed_data
```

```python
#taking a look at participants in each age category
imputed_data["Age_Category"].value_counts()
```

```python
#taking a look at different STD type variables
std_cols = [
    "STDs:condylomatosis", "STDs:cervical condylomatosis", "STDs:vaginal condylomatosis",
    "STDs:vulvo-perineal condylomatosis", "STDs:syphilis", "STDs:pelvic inflammatory disease",
    "STDs:genital herpes", "STDs:molluscum contagiosum", "STDs:AIDS", "STDs:HIV", "STDs:Hepatitis B", "STDs:HPV",
]
std_cols
```

```python
#grouping STD data by age category
imputed_data["total_std"] = imputed_data[std_cols].sum(axis=1)
std_agg = imputed_data.groupby("Age_Category", as_index=False)[std_cols].sum()
std_agg
```

```python
#looking at test columns -> tests are administered if cancer is suspected but only DX:Cancer is true diagnosis
test_cols = ["Hinselmann", "Schiller", "Citology", "Biopsy"]
imputed_data["total_tests"] = imputed_data[test_cols].sum(axis=1)

imputed_data.info()
```

```python
#trying to balance data with adasyn synthetic data generation
#otherwise dataset is definitely too small (not enough cancer diagnoses) for machine learning
#ADASYN looks at outliers and generates data from hard-to-learn examples
from imblearn.over_sampling import SMOTE, ADASYN
X = imputed_data.drop(["Dx:Cancer", "Age_Category"], axis=1)
y = imputed_data[["Dx:Cancer"]]
adasyn = ADASYN(random_state=42, sampling_strategy="minority")
x_adasyn, y_adasyn = adasyn.fit_resample(X, y)
imputed_data = x_adasyn.join(y_adasyn)
```

```python
#splitting the data
#chose stratification which allows each fold to have same number of ppl in specific age groups (strong correlation between
#age and predictor variables)
import sklearn
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
imputed_data["Age_Category"] = imputed_data["Age"].apply(age_category)
for train_idx, test_idx in split.split(imputed_data, imputed_data["Age_Category"]):
    train_set = imputed_data.loc[train_idx]
    test_set  = imputed_data.loc[test_idx]
cols_to_drop = ["Age_Category", "total_std", "total_tests"]
for dataset in (train_set, test_set):
    for col in cols_to_drop:
        dataset.drop(col, axis=1, inplace=True)

X_train = train_set.drop("Dx:Cancer", axis=1)
y_train = train_set["Dx:Cancer"].copy()
X_test  = test_set.drop("Dx:Cancer", axis=1)
y_test  = test_set["Dx:Cancer"].copy()
```

```python
#scaling the data and performing PCA via sklearn
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA

scaler = RobustScaler()
data_transform = scaler.fit_transform(X_train)
pca = PCA(n_components=0.99) #explained variance ratio 99%
data_pca = pca.fit_transform(data_transform) # fits and tranforms data
```

```python
#looking at explained variance ratio (y is like Eigen vectors explained variance ratio; x is Eigen vectors sorted by
#Eigen values) - get 13 total components
dimensions_bar = px.bar(x=np.arange(pca.n_components_), y=pca.explained_variance_ratio_)
dimensions_bar.show()
```

```python
#creating pipeline --> will use in Heroku app
#includes scaler and PCA
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("scaler", RobustScaler()),
    ("pca", PCA(n_components=13)) # components correspond to eigen vectors
])
X_train = pipeline.fit_transform(X_train)
X_test = pipeline.fit_transform(X_test)
```

```python
#running Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=500, max_leaf_nodes=None, max_features="sqrt")
rfc.fit(X_train, y_train)
preds = rfc.predict(X_test)

#printing rfc model metrics
from sklearn.metrics import classification_report

print(rfc.score(X_test, y_test)) #built in score is accuracy
print(classification_report(y_test, preds))
```

```python
#running logistic regression model
from sklearn.linear_model import LogisticRegression
params = {"C":np.logspace(-8,8, 15)}
log_reg = LogisticRegression(max_iter=1e7)

from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(log_reg, params, cv=10) #explored params and cv

clf.fit(X_train, y_train)
clf.best_score_
best_log_reg = clf.best_estimator_
best_log_reg.fit(X_train, y_train)
preds = best_log_reg.predict(X_test)
print(best_log_reg.score(X_test, y_test))
print(classification_report(y_test, preds))

```

```python
from sklearn.svm import SVC
svm = SVC()
params = {"C":np.logspace(-8,8, 15)}
clf = GridSearchCV(svm, params, cv=10)

clf.fit(X_train, y_train)
clf.best_score_
best_svm = clf.best_estimator_
preds = best_svm.predict(X_test)
print(best_svm.score(X_test, y_test))
print(classification_report(y_test, preds))
```

```python
#running KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
params = {"n_neighbors": np.arange(1,10)}
clf = GridSearchCV(knn, params, cv=10)

clf.fit(X_train, y_train)
clf.best_score_
best_knn = clf.best_estimator_
preds = best_knn.predict(X_test)
print(best_knn.score(X_test, y_test))
print(classification_report(y_test, preds))
```

```python
#running ANN

import tensorflow as tf
from tensorflow import keras
from keras import layers

tensor_data = tf.data.Dataset.from_tensor_slices((X_train, y_train))
tensor_batch = tensor_data.shuffle(1000).batch(32)

normalizer = keras.layers.Normalization()
normalizer.adapt(X_train)

model = keras.models.Sequential()
model.add(normalizer) #first layer that takes data and normalizes; scales data from 0 to 1 so that data will be on same scale
#and neural netowrk will weight each feature equally
model.add(layers.Dense(units=64, activation="relu")) #64 neurons in the layer activated by hinge function Relu on/0ff
model.add(layers.Dense(units=32, activation="relu"))
model.add(layers.Dense(units=32, activation="relu"))
model.add(layers.Dense(1)) #output layer

model.compile(optimizer="adam", loss=keras.losses.BinaryCrossentropy(from_logits=True), metrics=["accuracy"])
model.fit(tensor_batch, epochs=5)

preds = model.evaluate(X_test, y_test)

def sigmoid(x):
    return 1/(1 + np.exp(-x[0])) #logits
#print(list(map(sigmoid, np.round(model.predict(X_test), 2).tolist())))
```
